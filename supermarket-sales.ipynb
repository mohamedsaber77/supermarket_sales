   ## Data Cleaning



```python
# let's check if there are missing values or not
np.round(df.isnull().sum()*100.00/len(df),2)
```

```python
df.drop(['Invoice ID'], axis=1, inplace= True)
```

```python
# convert the date column to datetime 
df['Data'] = pd.to_datetime(df['Date'])
df['Time'] = pd.to_datetime(df['Time'])
```

```python
df.duplicated().sum()
```

```python
df['gross margin percentage'].unique()
```

```python
#Checking outliers of numerical data:
df_numeric_plot = df.drop(columns = ['Customer type', 'Gender', 'Product line', 'Date', 'Time', 'Rating'])
df_numeric_plot.plot(subplots=True, kind = 'box', layout = (4,4), figsize = (12,14), patch_artist=True)
plt.subplots_adjust(wspace = 0.5)
```

```python
def out_lairs(col):
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3-Q1
    data_out = df[((df[col] < (Q1-1.5*IQR)) | (df[col] > (Q3+1.5*IQR)))]
    return data_out[col]
out_lairs('Total')
```

```python
def imputing_outlairs(col):
    x=df[col].quantile(0.50)
    y=df[col].quantile(0.95)
    df[col] = np.where(df[col] >y, x, df[col])
```

```python
imputing_outlairs("Total")
```

```python
imputing_outlairs("cogs")
```

```python
imputing_outlairs("Tax 5%")
```

```python
imputing_outlairs("gross income")
```

```python
month_1 = df[df['month']==1]
month_2 = df[df['month']==2]
month_3 = df[df['month']==3]

months=['First Month', 'Second Month', 'Third Month'] 
fig, ax = plt.subplots(3,1, figsize=(10,20))

for i,month in enumerate([month_1, month_2, month_3]):
    fig.sca(ax[i])
    
    sns.histplot(data=month, x='gross income', color='#FF5733', kde=True)
    plt.title(months[i], fontsize=15, color='blue')
fig.suptitle('gross profit', fontsize=20, color='blue', y=.93)
 
                 
```
